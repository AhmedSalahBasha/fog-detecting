```python
def create_new_target(row):
    if row['Label1'] > 0 or row['Label2'] > 0:
        return 1
    else:
        return 0
```

--------------------------------------------------

```python
g07_left = pd.read_csv('../data/splitted_full_dataset/G07/trial_1/G07_Freezing_Trial1_trial_1_out_left_foot.csv')
g07_lower_left = pd.read_csv('../data/splitted_full_dataset/G07/trial_1/G07_Freezing_Trial1_trial_1_out_lower_left_foot.csv')
g07_right = pd.read_csv('../data/splitted_full_dataset/G07/trial_1/G07_Freezing_Trial1_trial_1_out_lower_right_foot.csv')
g07_lower_right = pd.read_csv('../data/splitted_full_dataset/G07/trial_1/G07_Freezing_Trial1_trial_1_out_right_foot.csv')

g07_left = g07_left.set_index('Time')
g07_lower_left = g07_lower_left.set_index('Time')
g07_right = g07_right.set_index('Time')
g07_lower_right = g07_lower_right.set_index('Time')

cols = ['Acc_1', 'Acc_2', 'Acc_3', 'Gyro_1', 'Gyro_2', 'Gyro_3', 'Label1', 'Label2']
g07_left = g07_left[[c for c in g07_left.columns if c in cols]]
g07_lower_left = g07_lower_left[[c for c in g07_lower_left.columns if c in cols]]
g07_right = g07_right[[c for c in g07_right.columns if c in cols]]
g07_lower_right = g07_lower_right[[c for c in g07_lower_right.columns if c in cols]]

g07_left['Label'] = g07_left.apply(lambda row: create_new_target(row), axis=1)
g07_left.drop(['Label1', 'Label2'], axis=1, inplace=True)

g07_left_1 = g07_left[g07_left['Label'] == 1]
g07_left_0 = g07_left[g07_left['Label'] == 0]
```

```python
g07_left[['Acc_1', 'Acc_2', 'Acc_3']].plot(figsize=(12, 6))
```

```python
g07_left.hist(figsize=(15, 10), bins=20, xlabelsize=8, ylabelsize=8)
```

```python
parameters = {'axes.labelsize': 14,
          'axes.titlesize': 16}
plt.rcParams.update(parameters)

fig, ax = plt.subplots(4,figsize=(15, 15), constrained_layout=True)
ax[0].plot(g07_left_1.index, g07_left_1[['Acc_1', 'Acc_2', 'Acc_3']])
ax[0].set_title('FoG', fontsize=13)
ax[0].legend(['Acc_X', 'Acc_Y', 'Acc_Z'])
ax[0].yaxis.grid(color='gray')
fig.suptitle('Accelerometer & Gyroscope Data - Left Shank Sensor', fontsize=16, fontweight ="bold",)

ax[1].plot(g07_left_0.index, g07_left_0[['Acc_1', 'Acc_2', 'Acc_3']])
ax[1].set_title('Non-FoG', fontsize=13)
ax[1].legend(['Acc_X', 'Acc_Y', 'Acc_Z'])
ax[1].yaxis.grid(color='gray')

ax[2].plot(g07_left_1.index, g07_left_1[['Gyro_1', 'Gyro_2', 'Gyro_3']])
ax[2].set_title('FoG', fontsize=13)
ax[2].legend(['Gyro_X', 'Gyro_Y', 'Gyro_Z'])
ax[2].yaxis.grid(color='gray')

ax[3].plot(g07_left_0.index, g07_left_0[['Gyro_1', 'Gyro_2', 'Gyro_3']])
ax[3].set_title('Non-FoG', fontsize=13)
ax[3].legend(['Gyro_X', 'Gyro_Y', 'Gyro_Z'])
ax[3].yaxis.grid(color='gray')

plt.xlabel('Time')
plt.savefig('../plots/acc_gyro_left_timeseries.png', dpi=100)
plt.show()
```
