```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed_data/full_dataset.csv', sep=',')
```

```python
df.head()
```

```python
df_fog = df[df['Label'] == 1]
df_nonfog = df[df['Label'] == 0]
```

```python
parameters = {'axes.labelsize': 14,
          'axes.titlesize': 16}
plt.rcParams.update(parameters)

fig, ax = plt.subplots(4,figsize=(15, 15), constrained_layout=True)
ax[0].plot(df_fog.index, df_fog[['Acc_1_left', 'Acc_2_left', 'Acc_3_left']])
ax[0].set_title('FoG', fontsize=13)
ax[0].legend(['Acc_X', 'Acc_Y', 'Acc_Z'])
ax[0].yaxis.grid(color='gray')

ax[1].plot(df_nonfog.index, df_nonfog[['Acc_1_left', 'Acc_2_left', 'Acc_3_left']])
ax[1].set_title('Non-FoG', fontsize=13)
ax[1].legend(['Acc_X', 'Acc_Y', 'Acc_Z'])
ax[1].yaxis.grid(color='gray')

ax[2].plot(df_fog.index, df_fog[['Gyro_1_left', 'Gyro_2_left', 'Gyro_3_left']])
ax[2].set_title('FoG', fontsize=13)
ax[2].legend(['Gyro_X', 'Gyro_Y', 'Gyro_Z'])
ax[2].yaxis.grid(color='gray')

ax[3].plot(df_nonfog.index, df_nonfog[['Gyro_1_left', 'Gyro_2_left', 'Gyro_3_left']])
ax[3].set_title('Non-FoG', fontsize=13)
ax[3].legend(['Gyro_X', 'Gyro_Y', 'Gyro_Z'])
ax[3].yaxis.grid(color='gray')

fig.suptitle('Accelerometer & Gyroscope Data - Left Shank Sensor', fontsize=16, fontweight ="bold",)
plt.xlabel('Time')
plt.savefig('../plots/acc_gyro_left_timeseries_fulldataset.png', dpi=100)
plt.show()
```
