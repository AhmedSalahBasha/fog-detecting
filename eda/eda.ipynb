```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed_data/full_dataset.csv', sep=',')
df.describe(percentiles=[.05, .20, 0.50, .80, .95])
```

```python
new_df = df.copy()
gyro_x_df = new_df.filter(regex=('Gyro_1'))
gyro_x = pd.concat([gyro_x_df['Gyro_1_left'],
                    gyro_x_df['Gyro_1_right'],
                    gyro_x_df['Gyro_1_lowerleft'],
                    gyro_x_df['Gyro_1_lowerright']], ignore_index=True)
gyro_x.rename('gyro_x', inplace=True)
gyro_y_df = new_df.filter(regex=('Gyro_2'))
gyro_y = pd.concat([gyro_y_df['Gyro_2_left'],
                    gyro_y_df['Gyro_2_right'],
                    gyro_y_df['Gyro_2_lowerleft'],
                    gyro_y_df['Gyro_2_lowerright']], ignore_index=True)
gyro_y.rename('gyro_y', inplace=True)
gyro_z_df = new_df.filter(regex=('Gyro_3'))
gyro_z = pd.concat([gyro_z_df['Gyro_3_left'],
                    gyro_z_df['Gyro_3_right'],
                    gyro_z_df['Gyro_3_lowerleft'],
                    gyro_z_df['Gyro_3_lowerright']], ignore_index=True)
gyro_z.rename('gyro_z', inplace=True)
acc_x_df = new_df.filter(regex=('Acc_1'))
acc_x = pd.concat([acc_x_df['Acc_1_left'],
                    acc_x_df['Acc_1_right'],
                    acc_x_df['Acc_1_lowerleft'],
                    acc_x_df['Acc_1_lowerright']], ignore_index=True)
acc_x.rename('acc_x', inplace=True)
acc_y_df = new_df.filter(regex=('Acc_2'))
acc_y = pd.concat([acc_y_df['Acc_2_left'],
                    acc_y_df['Acc_2_right'],
                    acc_y_df['Acc_2_lowerleft'],
                    acc_y_df['Acc_2_lowerright']], ignore_index=True)
acc_y.rename('acc_y', inplace=True)
acc_z_df = new_df.filter(regex=('Acc_3'))
acc_z = pd.concat([acc_z_df['Acc_3_left'],
                    acc_z_df['Acc_3_right'],
                    acc_z_df['Acc_3_lowerleft'],
                    acc_z_df['Acc_3_lowerright']], ignore_index=True)
acc_z.rename('acc_z', inplace=True)
```

```python
frames = [gyro_x, gyro_y, gyro_z, acc_x, acc_y, acc_z]
combo_df = pd.concat(frames, axis=1, keys=[s.name for s in frames])
combo_df.head()
```

```python
stats_df = combo_df.describe(percentiles=[.05, .20, 0.50, .80, .95])
stats_df = stats_df[stats_df.columns.to_list()].round(3)
```

```python
stats_df
```

```python
axarr = combo_df.hist(figsize=(15, 10), bins=50, xlabelsize=14, ylabelsize=14, layout=(2, 3), sharey=True)
for ax in axarr.flatten():
    ax.set_xlabel("Sensor Values")
    ax.set_ylabel("Frequency")
plt.savefig('../plots/df_hist.png', dpi=100)
plt.show()
```
